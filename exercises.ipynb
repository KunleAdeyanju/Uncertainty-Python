```python
import random

def sim_event(x):
    # so it has to also equal the number cause 100% 
    if random.random() <= (x): 
        return "foo"
    else:
        return "bar"
```

```python
#sim_event(float(2/7))
number_of_repetitions = 100000
samples = [sim_event(2/7) for repetition in range(number_of_repetitions)]
samples
```
